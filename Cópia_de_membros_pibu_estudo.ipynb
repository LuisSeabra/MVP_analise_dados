Distribuição de Frequências

```python
import pandas as pd

# Carregar o arquivo Excel
df = pd.read_excel('membros_pibu.xlsx')

# 1. Definir intervalos de 10 anos
bins = range(0, df['Idade'].max() + 10, 10)  # Cria intervalos até a idade máxima, com passo de 10

# 2. Categorizar a coluna 'Idade' nos intervalos definidos
df['Faixa_Etaria'] = pd.cut(df['Idade'], bins=bins)

# 3. Frequência simples
frequencia_simples = df['Faixa_Etaria'].value_counts().sort_index()

# 4. Frequência relativa
frequencia_relativa = frequencia_simples / frequencia_simples.sum()

# 5. Frequência acumulada
frequencia_acumulada = frequencia_simples.cumsum()

# 6. Frequência relativa acumulada
frequencia_relativa_acumulada = frequencia_relativa.cumsum()

# Criar um DataFrame com todas as frequências
distribuicao_frequencias = pd.DataFrame({
    'Frequência Simples': frequencia_simples,
    'Frequência Relativa (%)': frequencia_relativa * 100,
    'Frequência Acumulada': frequencia_acumulada,
    'Frequência Relativa Acumulada (%)': frequencia_relativa_acumulada * 100
})

# Exibir a tabela de distribuição de frequências
distribuicao_frequencias
```

```python
# @title Texto de título padrão
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar o arquivo Excel
df = pd.read_excel('membros_pibu.xlsx')

# Verificar e tratar valores ausentes
df['Idade'] = pd.to_numeric(df['Idade'], errors='coerce')

# Resumo estatístico
resumo_estatistico = df['Idade'].describe()

# Definir faixas etárias
bins = range(0, df['Idade'].max() + 10, 10)
df['Faixa_Etaria'] = pd.cut(df['Idade'], bins=bins)

# Gráficos
plt.figure(figsize=(15, 12))

# Gráfico 1: Gráfico de Barras
plt.subplot(3, 2, 1)
faixa_etaria_counts = df['Faixa_Etaria'].value_counts().sort_index()
faixa_etaria_counts.plot(kind='bar', color='skyblue')  # Cor: skyblue
plt.title('Distribuição de Idades por Faixa Etária')
plt.xlabel('Faixa Etária')
plt.ylabel('Frequência')

# Gráfico 2: Histograma
plt.subplot(3, 2, 2)
df['Idade'].plot(kind='hist', bins=bins, color='green', edgecolor='black')  # Cor: green
plt.title('Histograma da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')

# Gráfico 3: Gráfico de Distribuição (KDE)
plt.subplot(3, 2, 3)
sns.kdeplot(df['Idade'], fill=True, color='purple')  # Cor: purple, usando fill=True
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Densidade')

# Gráfico 4: Boxplot da Idade com rótulos
plt.subplot(3, 2, 4)
box = sns.boxplot(y=df['Idade'], color='yellow')  # Boxplot em yellow
plt.title('Boxplot da Idade')
plt.ylabel('Idade')

# Adicionando rótulos para mediana e quartis
mediana = df['Idade'].median()
q1 = df['Idade'].quantile(0.25)
q3 = df['Idade'].quantile(0.75)

# Adicionar rótulos
plt.text(0, mediana, f'Mediana: {mediana:.1f}', ha='center', va='bottom', fontsize=10, color='black')
plt.text(0, q1, f'Q1: {q1:.1f}', ha='center', va='bottom', fontsize=10, color='blue')
plt.text(0, q3, f'Q3: {q3:.1f}', ha='center', va='bottom', fontsize=10, color='blue')

# Adicionando rótulo do mínimo e máximo
minimo = df['Idade'].min()
maximo = df['Idade'].max()
plt.text(0, minimo, f'Mín: {minimo:.1f}', ha='center', va='bottom', fontsize=10, color='green')
plt.text(0, maximo, f'Máx: {maximo:.1f}', ha='center', va='bottom', fontsize=10, color='red')

# Gráfico 5: Gráfico de Pizza na última posição
plt.subplot(3, 2, 5)
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)  # Ajuste de margens
frequencia_relativa = faixa_etaria_counts / faixa_etaria_counts.sum() * 100
plt.pie(frequencia_relativa, labels=frequencia_relativa.index,
        autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors,
        textprops={'fontsize': 10})  # Tamanho de fonte dos rótulos
plt.title('Distribuição Relativa das Idades por Faixa Etária')
plt.axis('equal')

# Mostrar todos os gráficos
plt.tight_layout()
plt.show()

# Exibir resumo estatístico
print("Resumo Estatístico da Idade:")
print(resumo_estatistico)
```

Distribuição de fequencias Estado Civil por faixa etária

```python
import pandas as pd

# Carregar o arquivo Excel
df = pd.read_excel('membros_pibu.xlsx')

# Verificar e tratar valores ausentes na coluna Idade e EstadoCívil
df['Idade'] = pd.to_numeric(df['Idade'], errors='coerce')
df['EstadoCívil'] = df['EstadoCívil'].fillna('Desconhecido')  # Substituir valores ausentes

# Definir faixas etárias
bins = range(0, df['Idade'].max() + 10, 10)  # Faixas de 10 em 10 anos
labels = [f'{i}-{i + 9}' for i in bins[:-1]]  # Criar rótulos para as faixas
df['Faixa_Etaria'] = pd.cut(df['Idade'], bins=bins, labels=labels, right=False)

# Calcular a distribuição de EstadoCívil por faixa etária
distribuicao_estado_civil = df.groupby(['Faixa_Etaria', 'EstadoCívil']).size().unstack(fill_value=0)

# Exibir a distribuição
print(distribuicao_estado_civil)
```
